```python
import re
```

```python
def clean_gender(x):
    dicc_gender = {"M":re.search("^M",str(x)),"F":re.search("^W",str(x))}
    for key,values in dicc_gender.items():
        if values:
            return key
    return "other"
```

```python
df_cross_games["gender"]=df_cross_games["division"].apply(clean_gender)
```
